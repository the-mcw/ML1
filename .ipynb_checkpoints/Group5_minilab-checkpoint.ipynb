# Logistic Regression

```python
#LogReg using GridSearch
X = drivers_6[['pointsSeason','qualifyPosition','minQ','positionOrder','laps','rank','mean_stop','stops','round','age','ConstructorSeasonPoints']]
Y = drivers_6['nextPodium']

#encoder = OneHotEncoder()
#X_encoded = encoder.fit_transform(X)

encoder = LabelEncoder()
Y_encoded = encoder.fit_transform(Y)

X_train, X_test, y_train, y_test = train_test_split(X, Y_encoded, test_size=0.2, random_state=5)

scl_obj = StandardScaler()
scl_obj.fit(X_train)

X_train_scaled = scl_obj.transform(X_train) 
X_test_scaled = scl_obj.transform(X_test)
```

```python
parameters = {
    'penalty': ['l1', 'l2'],
    'C': [0.1, 1, 10, 100],
    'solver': ['liblinear']
}

LogReg = LogisticRegression(max_iter=1000)

gridsearch_LR = GridSearchCV(estimator=LogReg, param_grid=parameters, cv=5)
gridsearch_LR.fit(X_train_scaled, y_train)

LogReg_model = gridsearch_LR.best_estimator_
LogReg_params = gridsearch_LR.best_params_
```

```python
#Predictions
LogReg_y_pred = LogReg_model.predict(X_test_scaled)
```

```python
#Accuracy
accuracy = accuracy_score(y_test, LogReg_y_pred)

print("Best Accuracy:", accuracy)
print("Best Parameters:", LogReg_params)
```

```python
weights_LR = LogReg_model.coef_
print(weights_LR)

y_intercept = LogReg_model.intercept_
print(y_intercept)
```

```python
weights_LR = pd.Series(weights_LR[0],index=X.columns)
weights_LR.plot(kind='bar')
plt.show()
```

The importance of each variable is determined by the weight. Variables with higher absolute weights have a stronger influence on the predicted outcome. For the Logistic Regression model, 'pointsSeason' and 'qualifyPosition' have the highest absolute weights, suggesting that they are important factors in determining podium finishes. Weight of 0.8966 for 'pointsSeason' suggests that an increase in the driver's points in the current season as of that particular race is positively correlated to the likelihood of the driver ending up on the podium. On the contrary, weight of -1.6429 for 'qualifyPosition' indicates that a higher qualifying position (ex. positions 10 - 20, farther down on the starting grid) reduces the chances of the driver getting a podium finish.

# SVM

```python
# SVM Grid Search

parameter_svm = {
    'C': [0.1, 1, 10, 100, 1000],
    'kernel': ['linear', 'rbf']
}

SVM = SVC()
```

```python
gridsearch_svm = GridSearchCV(SVM, param_grid = parameter_svm, refit = True, verbose = 3)
gridsearch_svm.fit(X_train_scaled, y_train)

SVM_model = gridsearch_svm.best_estimator_
SVM_params = gridsearch_svm.best_params_
```

```python
# Predictions
SVM_pred = SVM_model.predict(X_test_scaled)
```

```python
# Accuracy
SVM_accuracy = accuracy_score(y_test, SVM_pred)

print("Best Accuracy:", SVM_accuracy)
print("Best Parameters:", SVM_params)
```

```python
print(SVM_model.support_vectors_.shape)
print(SVM_model.support_.shape)
print(SVM_model.n_support_)
```

```python
print(SVM_model.coef_)
weights_SVM = pd.Series(SVM_model.coef_[0],index=X.columns)
weights_SVM.plot(kind='bar')
```

```python
X_support = X_train.iloc[SVM_model.support_,:]

X_support['podium'] = y_train[SVM_model.support_,]

X_support.info()
```

```python
# group the original data and the support vectors
X_grouped_support = X_support.groupby(['podium'])
X_grouped = drivers_6.groupby(['nextPodium'])

# plot KDE of Different variables
vars_to_plot = ['qualifyPosition','pointsSeason','round','rank','mean_stop']

for v in vars_to_plot:
    plt.figure(figsize=(10,5))
    # plot support vector stats
    plt.subplot(1,2,1)
    ax = X_grouped_support[v].plot.kde() 
    plt.legend(['Podium','Missed'])
    plt.title(v+' (Instances chosen as Support Vectors)')
    
    # plot original distributions
    plt.subplot(1,2,2)
    ax = X_grouped[v].plot.kde() 
    plt.legend(['Podium','Missed'])
    plt.title(v+' (Original)')

```

```python
sns.jointplot(data=X_support, x="qualifyPosition", y="pointsSeason", hue="podium", kind='kde')

sns.jointplot(data=drivers_6, x="qualifyPosition", y="pointsSeason", hue="nextPodium", kind='kde')
```

```python
sns.jointplot(data=X_support, x="round", y="pointsSeason", hue="podium", kind='kde')

sns.jointplot(data=drivers_6, x="round", y="pointsSeason", hue="nextPodium", kind='kde')
```

Looking at the support vectors, the differences in distributions for podium finishers and non-podium finishers is significantly smaller than in the whole dataset. This makes sense, since SVM's really focus in on the boundary line and where instances are hard to distringuish from each other, as those that are already easily separable will remain that way.

# Logistic Regression vs. Support Vector Machine

Results suggest that the SVM model slightly outperformed the Logistic Regression Model with an accuracy of 0.9125. Parameters for the best SVM model are C=10 and kernel='linear'. Overall, both models achieved high accuracy, but in terms of training time, the logistic regression model was relatively faster to train compared to the SVM model. SVMs can take longer to train, especially when using linear kernel models.  
To note, the independent variables for both models have been standardized to ensure fair comparisons between features with different scales/ranges.  
Logistic regression is easy to understand and faster to compute. It works well when the decision boundary is linear. On the other hand, SVMs are more flexible and can handle complicated data. They are good at finding non-linear patterns and can give better accuracy when the relationship between features and the predicting variable is complex.

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, make_scorer, confusion_matrix, ConfusionMatrixDisplay

X = drivers_6[['pointsSeason','qualifyPosition','minQ','positionOrder','laps','rank','mean_stop','stops','round','age','ConstructorSeasonPoints']]
pod = drivers_6['nextPodium']
top = drivers_6['nextTop10']

encoder = LabelEncoder()
pod_encoded = encoder.fit_transform(-pod)
top_encoded = encoder.fit_transform(-top)

X_train, X_test, pod_train, pod_test, top_train, top_test = train_test_split(X, pod_encoded, top_encoded, test_size=0.2, random_state=5)

scl_obj = StandardScaler()
scl_obj.fit(X_train)

X_train_scaled = scl_obj.transform(X_train) 
X_test_scaled = scl_obj.transform(X_test)
```

```python
knn = KNeighborsClassifier()
param_knn = {
    'n_neighbors': [3, 5, 7, 9, 11]
}
scoring = {'AUC': 'roc_auc', 'Precision': make_scorer(precision_score)}

podKNN_gs = GridSearchCV(knn,
                      param_grid=param_knn,
                      scoring=scoring,
                      refit='Precision',
                      cv=10,
                      verbose=3)
```

## Podium

```python
podKNN_gs.fit(X_train_scaled, pod_train)

knnPodium_model = podKNN_gs.best_estimator_
knnPodium_params = podKNN_gs.best_params_
```

```python
knnPodium_params
```

```python
knnPodiumPred = knnPodium_model.predict(X_test_scaled)
ConfusionMatrixDisplay(confusion_matrix(pod_test,knnPodiumPred),
                       display_labels={'Podium Finish','Non-Podium Finish'}).plot()
plt.tick_params(axis=u'both', which=u'both',length=0)
plt.grid(visible=None)
plt.show()
```

## Top 10

```python
param_knn = {
    'n_neighbors': [3, 5, 7, 9, 11]
}

topKNN_gs = GridSearchCV(knn,
                      param_grid=param_knn,
                      scoring=scoring,
                      refit='Precision',
                      cv=10,
                      verbose=3)
```

```python
topKNN_gs.fit(X_train_scaled, top_train)

knnTop10_model = topKNN_gs.best_estimator_
knnTop10_params = topKNN_gs.best_params_
```
